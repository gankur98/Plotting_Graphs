```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Pie Graphs
sizes = [3, 4, 6, 2]
colors = ['red', 'pink', 'green', 'blue']
explode = [0.2, 0, 0, 0]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, explode = explode, colors = colors, labels = labels, autopct = '%.2f%%')
plt.title("Split among Classes")
plt.axis("equal")
plt.show()
```

```python
#Pie Graphs
sizes = [65, 30, 5]
labels = ['Male', 'Female', 'Other']
explode = [0, 0, 0.3]
plt.pie(sizes, labels = labels, explode = explode, autopct='%.2f%%')
plt.title("Split among Gender")
plt.show()
```

```python
#Line Graphs
x = np.arange(0, 5, 0.1)
y1 = x**3
y2 = x**2
plt.plot(x, y1, linewidth=5, color='Blue', label = 'x^3')
plt.plot(x, y2, linewidth=5, color='Red', label = 'x^2')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Demo of X^3 and X^2')

plt.grid()
plt.show()
```
